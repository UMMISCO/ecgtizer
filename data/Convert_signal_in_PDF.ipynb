```python
import sys
sys.path.append("../ecgtizer/")
from XML2PDF import Write_PDF
import os
import pandas as pd

import numpy as np
```

```python
# Specify the directory path
directory = "PTB-XL/Original/"

# List all the files in the directory
files = os.listdir(directory)

# Print the list of files
for file in files:

    file_path = directory + file
    df = pd.read_csv(file_path)

    # Transform the DataFrame into a dictionary
    ecg = df.to_dict()

    for k in ecg.keys():
        # The signal is originally in µV and needs to be converted to mV to match the scale of the paper format.
        ecg[k] = np.array(list((ecg[k].values())))*1000

    Write_PDF (ecg, "PTB-XL/PDF/"+file.split(".")[0]+'.pdf', "type1", lead_IIc = ecg["II"])


```
